## 課題1

### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
for i in range(5):
    print ("*" * ( i + 1))
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5
for i in range(size):
    print (" " * (i * 1) + "*")
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5
for i in range(size):
    print (" " * (size - i - 1) + "*")
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5
for i in range(size):
    for j in range(size):
        if j == i or j == size - 1 - i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5
for i in range(size):
    print (" " * (size - i - 1) + "*" * (2 * i + 1))
````
